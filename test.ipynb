```python
td = np.datetime64("2026-06-29T14:30:15.123456") - np.datetime64("2026-06-28T12:30:15.123456")
print(td.astype("timedelta64[s]"))
print(td.astype("timedelta64[ns]").astype(int)/1e9)

print(60e9)
```

```python
def td_to_iso(td):
    # Computes np.timedelta64 to ISO8601 standard
    # cannot handle variable-length units such as months and years
    
    nanoseconds = td.astype("timedelta64[ns]").astype(int)
    
    minutes, seconds = divmod(nanoseconds, 60e9)
    hours, minutes = divmod(minutes, 60)
    days, hours = divmod(hours, 24)

    # convert microseconds to fractional seconds
    microseconds = str(td.microseconds/1e6)[1:]

    return f"P{days}D{hours}H{minutes}M{seconds}S"

td_to_iso(td)

print(isinstance(td, timedelta))
```

```python

dt1 = datetime.fromisoformat(
    "2026-06-29T14:30:15+02:00"
)

dt2 = datetime.fromisoformat(
    "2026-06-29T15:33:15+02:00"
)

dt = dt2-dt1
print(td_to_iso(dt))
```
